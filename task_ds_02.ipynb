```python
# Cell 1: imports + load files (works if you uploaded train.csv/test.csv)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline

# Try loading files from current folder (uploaded in Colab)
train_path = "train.csv"
test_path = "test.csv"

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)

print("Train shape:", train.shape)
print("Test shape :", test.shape)

# show first 5 rows so you see the data
display(train.head())
```

```python
# Cell 2: quick info + missing values
print("Columns:", train.columns.tolist())
print("\nTrain info:")
display(train.info())
print("\nMissing values (train):")
display(train.isnull().sum().sort_values(ascending=False))

print("\nSample statistics (numeric):")
display(train.describe().T)
```

```python
# Cell 3: target distribution (Survived)
plt.figure(figsize=(6,4))
ax = sns.countplot(x='Survived', data=train)
ax.set_xticklabels(['Did not survive (0)','Survived (1)'])
ax.set_title('Survival counts')
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+p.get_width()/2., p.get_height()),
                ha='center', va='bottom')
plt.show()

# Percent survived
pct_survived = train['Survived'].mean() * 100
print(f"Percent survived: {pct_survived:.2f}%")
```

```python
# Cell 4a: Survival by Sex
plt.figure(figsize=(8,4))
sns.countplot(x='Sex', hue='Survived', data=train)
plt.title('Survival counts by Sex (male/female)')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.show()

# Cell 4b: Survival rate by Pclass
pclass_tab = pd.crosstab(train['Pclass'], train['Survived'], normalize='index') * 100
display(pclass_tab.round(2))

plt.figure(figsize=(6,4))
pclass_tab.plot(kind='bar', stacked=True)
plt.title('Survival percentage by Passenger Class (Pclass)')
plt.ylabel('Percentage (%)')
plt.xlabel('Pclass (1 = highest)')
plt.show()
```

```python
# Cell 5a: Age distribution (histogram + KDE)
plt.figure(figsize=(8,4))
sns.histplot(train['Age'].dropna(), bins=30, kde=True)
plt.title('Age distribution (all passengers)')
plt.xlabel('Age')
plt.show()

# Cell 5b: Age distribution separated by survival (KDE)
plt.figure(figsize=(8,4))
sns.kdeplot(train.loc[train['Survived']==0,'Age'].dropna(), label='Did not survive', shade=True)
sns.kdeplot(train.loc[train['Survived']==1,'Age'].dropna(), label='Survived', shade=True)
plt.title('Age distribution by Survival')
plt.xlabel('Age')
plt.legend()
plt.show()
```

```python
# Cell 6: Fare distribution & boxplot by Pclass
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(train['Fare'].dropna(), bins=30)
plt.title('Fare distribution')

plt.subplot(1,2,2)
sns.boxplot(x='Pclass', y='Fare', data=train)
plt.title('Fare by Passenger Class')
plt.show()
```

```python
# Cell 7: Embarked and Cabin summary
print("Embarked value counts:")
display(train['Embarked'].value_counts(dropna=False))

print("\nCabin missing count:", train['Cabin'].isnull().sum())
print("Total rows:", len(train))
```

```python
# Cell 8: cleaning - explainable & simple approach
clean = train.copy()

# 1) Drop 'Cabin' because >75% missing in train (too many to reliably impute)
clean = clean.drop(columns=['Cabin'])

# 2) Fill 'Embarked' missing values with the mode (most common port)
emb_mode = clean['Embarked'].mode()[0]
clean['Embarked'] = clean['Embarked'].fillna(emb_mode)

# 3) Fill 'Age' missing values with median (robust to outliers)
age_median = clean['Age'].median()
clean['Age'] = clean['Age'].fillna(age_median)

# 4) If any missing 'Fare' in test later, we'll fill with median (example)
fare_median = clean['Fare'].median()
clean['Fare'] = clean['Fare'].fillna(fare_median)

print("Missing values after cleaning (train):")
display(clean.isnull().sum())
```

```python
# Cell 9: create family_size and is_alone, extract title from Name
df2 = clean.copy()
df2['FamilySize'] = df2['SibSp'] + df2['Parch'] + 1   # +1 for the passenger
df2['IsAlone'] = (df2['FamilySize'] == 1).astype(int)

# Extract title from Name (Mr, Mrs, Miss, Master, etc.)
df2['Title'] = df2['Name'].str.extract(r',\s*([^\.]+)\.')
df2['Title'] = df2['Title'].str.strip()
# Group rare titles
title_counts = df2['Title'].value_counts()
rare_titles = title_counts[title_counts < 10].index
df2['Title'] = df2['Title'].replace(rare_titles, 'Other')

display(df2[['Name','Title','FamilySize','IsAlone']].head())
```

```python
# Convert categorical to numeric for correlation
corr_df = df2.copy()
corr_df['Sex'] = corr_df['Sex'].map({'male':0,'female':1})
corr_df['Embarked'] = corr_df['Embarked'].map({'S':0,'C':1,'Q':2})

# pick numeric features to check correlation with Survived
cols = ['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','FamilySize','IsAlone']
plt.figure(figsize=(8,6))
sns.heatmap(corr_df[cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

```python
# Survival rate by Title
title_surv = df2.groupby('Title')['Survived'].mean().sort_values(ascending=False)
display((title_surv*100).round(2))

plt.figure(figsize=(8,4))
sns.barplot(x=title_surv.index, y=title_surv.values*100)
plt.ylabel('Survival rate (%)')
plt.title('Survival rate by Title')
plt.show()

# Survival rate by FamilySize
fs_surv = df2.groupby('FamilySize')['Survived'].mean()
plt.figure(figsize=(8,4))
sns.lineplot(x=fs_surv.index, y=fs_surv.values*100, marker='o')
plt.xlabel('Family Size')
plt.ylabel('Survival rate (%)')
plt.title('Survival rate vs Family Size')
plt.show()
```
